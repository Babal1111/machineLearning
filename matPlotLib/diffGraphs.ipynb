```python
import matplotlib.pyplot as plt

x= ["swif","tiago","nexon","innova"]
y= [100,200,400,200]
plt.bar(x,y)
plt.show()
```

```python
regions = ["north","south","east","west"]
revenue = [1000,2000,1500,1000]

plt.pie(revenue, labels =regions,autopct= '%.1f%%',colors = ['gold','pink','red','green'])
plt.title("paddy contribution")
plt.show()
```
