```python
 import matplotlib.pyplot as plt

hrs_studied = [1,2,3,4,5,6,7,8]
scores = [50,55,60,65,70,80,90,95]

plt.scatter(hrs_studied,scores,color = 'green', marker='o', label = 'Student Data')
plt.xlabel('Hours studied')
plt.ylabel('Marks obtained')
plt.show()
```
