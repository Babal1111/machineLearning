```python
import matplotlib.pyplot as plt

y = [10,30,40,20,60]
x=['a','b','c','d','e']

plt.plot(x,y, color = "red", linestyle = '--',linewidth = 2, marker = 'o', label = "sales  in 2025")
plt.xlabel("months")
plt.ylabel("sales")
plt.title("Babal's boutique")
plt.legend()
plt.grid(color = "blue",linewidth= 1)
plt.xlim
plt.show()

```
