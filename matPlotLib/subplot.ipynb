```python
import matplotlib.pyplot as plt

x =[1,2,3,4,5]
y = [10,20,40,25,30]

plt.subplot(1,2,1) #1st row 2nd col 1st subplot
plt.plot(x,y)
plt.title("line graph")

plt.subplot(1,2,2)#1st row 2nd col 1st subplot
plt.bar(x,y)
plt.title("bar graph")

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(1,2,figsize = (10,5))
x =[1,2,3,4,5]
y = [10,20,40,25,30]

ax[0].plot(x,y)
ax[0].set_title("line graph")

ax[1].bar(x,y,color="green")
ax[1].set_title("bar graph")

fig.suptitle("comparision of diff bars")

plt.tight_layout()
plt.show()
```
